```python
# import the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
%matplotlib inline
WData = pd.read_csv("C:\\Users\\oakin\\Documents\\Data Science Udemy\\Python Programming\\Unit Anaytics\\rawdata2.csv")
WData
```

```python
# We need to remove the column names so that the data frame dosn't act like a contigency table
x = len(WData.columns)
WData.columns = np.arange(x)
WData
```

```python
WData  = WData.transpose()
WData
```

```python
# rename column names and drpping unwanted rows 
WData.columns = ['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday']
WData = WData.drop(0,0)
```

```python
WData = WData.drop(5,0)
WData
```

```python
#converting data to integer type. it took object type because of the contingency table attribute mentioned earlier 
WData = WData.astype('int')
WData.info()
```

```python
WData.describe().transpose()
```

```python
#Creating a field for the fiscal years 
WData.insert (7, "FYear",['FY2018/19','FY2019/20','FY2020/21','FY2021/22'])
WData
```

```python
# Average plot by DoW- barchart
p =sns.barplot(data = WData)
```

```python
import scipy.stats as sc
value = sc.f_oneway(WData['Sunday'], WData['Monday'], WData['Tuesday'], WData['Wednesday'], WData['Thursday'], WData['Friday'], WData['Saturday'] )
if value[1] <= 0.05:
    print('Reject Null Hypothesis', ",p-vlaue=", value[1])
else: 
    print('Accept Null Hypothesis',",p-vlaue=", value[1])
```
